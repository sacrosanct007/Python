The below code imports a salary data, performs SLR and visualizes the result

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#importing datasets
data=pd.read_csv('Salary_Data.csv')
data.head()
```

```python
X=data.iloc[:,:-1].values
Y=data.iloc[:,-1].values
Y[0]
```

```python
#Splitting dataset into training and testing datasets
from sklearn.cross_validation import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#Feature scaling is not required for Simple linear regression
#Fitting SLR to training set
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# predicting the test set results
Y_pred=regressor.predict(X_test)
```

```python
# Visualizing the training set results
# Note that points are plotted using scatter graph and line via plot
plt.scatter(X_train,Y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Salary vs experience(Training Set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualizing the test set results
plt.scatter(X_test,Y_test,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Salary vs experience(Test Set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
